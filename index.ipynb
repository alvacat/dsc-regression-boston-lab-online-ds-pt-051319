### Let's do Regression 

Now, let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). You'll do this in a loop and in every iteration, you should pick one of the independent variables. Perform the following steps:

* Run a simple OLS regression between independent and dependent variables
* Plot a regression line on the scatter plots
* Plot the residuals using `sm.graphics.plot_regress_exog()`
* Plot a Q-Q plot for regression residuals normality test 
* Store following values in array for each iteration: <-- Still need to do this bullet point
    * Independent Variable
    * r_squared'
    * intercept'
    * 'slope'
    * 'p-value'
    * 'normality (JB)' 
* Comment on each output 

```python
import scipy.stats as stats
sns.set_style("darkgrid")
regression_info=[['independent variable','r squared','intercept','slope','p value','normality (JB)']]

for column in boston_subset:
    if column!="medv":
        f='medv~'+column
        model=ols(formula=f,data=boston_subset).fit()
        new_x=pd.DataFrame({column:[boston_subset[column].min(),boston_subset[column].max()]})
        new_y=model.predict(new_x)

        plt.scatter(boston_subset[column],boston_subset['medv'],color="blue",label=column)
        plt.plot(new_x,new_y,color="red",label="regression model")
        plt.title(column)
        plt.show()
        
        fig=plt.figure(figsize=(15,8))
        fig=sm.graphics.plot_regress_exog(model, column, fig=fig)
        plt.show()

        residuals = model.resid
        fig=sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
        plt.show()
        
        regression_info.append([column,
                                model.rsquared,
                                model.params[0], 
                                model.params[1], 
                                model.pvalues[1], 
                                sm.stats.jarque_bera(model.resid)[0]])
    else:
        continue
```

```python
pd.DataFrame(regression_info)
```

### Your observations here 
The r squared values are all pretty small which indicates that the models aren't very reliable.  The intercepts and slopes don't tell us much about the reliability.  The normality numbers are all way higher than 6 so none of these are normally distributed, which means they don't meet the requirements for an OLS model (and so the p value and r-squared don't really matter).

Clearly, the results are not very reliable. The best R-Squared is witnessed with `rm`, so in this analysis, this is your best predictor.
